#1.Importando as bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

```

#2.Importando o dataset e medidas iniciais

```python
data = pd.read_csv('/content/Shill_Bidding_Dataset.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

#3.Tratamento dos dados

```python
# remove the columns that are irrelevant to the case study

data.drop(["Record_ID","Auction_ID","Bidder_ID"],axis=1,\
inplace=True)
```

```python
data.head()
```

#4.Desenvolvendo o modelo

```python
X = data.drop("Class",axis=1)
y = data["Class"]

# Split the dataset into training and testing sets

X_train,X_test,y_train,y_test = train_test_split\
(X,y,test_size=0.3,\
random_state=1, stratify=y)
```

```python
# Print the size of training and testing datasets

print("Training dataset size: {}, Testing dataset size: {}"\
.format(X_train.shape,X_test.shape))
```

```python
# Fit the model using linear regression

linear = linear_model.LinearRegression()
linear.fit(X_train,y_train)
```

```python
# Predict on the first 10 test data points using the following code

linear.predict(X_test)[:10]
```

```python
# Now check the actual target values using y_test[:10].values:
y_test[:10].values
```

```python
# Evaluate the score of the linear regression model on the training and testing datasets:

print("Score on training dataset: {}, "\
"Score on testing dataset: {}"\
.format(linear.score(X_train,y_train),\
linear.score(X_test,y_test)))
```

```python
# Fit the model using logistic regression as follows

logit = linear_model.LogisticRegression()
logit.fit(X_train,y_train)
```

```python
# Predict on the test data:

logit.predict(X_test)[:10]
```

```python
# Check the actual target values with y_test[:10].values:

y_test[:10].values
```

```python
# Similar to linear regression, find the score on the training and testing datasets as follows:

print("Score on training dataset: {}, "\
"Score on testing dataset: {}"\
.format(logit.score(X_train,y_train),\
logit.score(X_test,y_test)))
```
